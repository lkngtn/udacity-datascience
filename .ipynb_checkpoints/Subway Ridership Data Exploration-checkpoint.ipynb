Now we will import our dataset and confirm that the columns we see in our dataframe are what we expect. 

```python
df = pd.read_csv('data/improved-dataset/turnstile_weather_v2.csv')
print df.columns
```

Lets get some basic summary information about our dataset. 

```python
print "Average by Hour:"
entries_by_time = df.groupby('TIMEn')['ENTRIESn_hourly'].mean()

entries_by_time.plot(kind='bar')
```

```python
df.boxplot('ENTRIESn_hourly','DATEn')
```

```python
avg_hourly_entries_by_time = df.groupby('TIMEn')['ENTRIESn_hourly']
print avg_hourly_entries_by_time.mean()

print "Weekdays Only"
print df[df['weekday'] == 0].groupby('TIMEn')['ENTRIESn_hourly'].mean()

print "Weekends Only"
print df[df['weekday'] == 1].groupby('TIMEn')['ENTRIESn_hourly'].mean()
```
